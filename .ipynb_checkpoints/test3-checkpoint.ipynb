```python
requirement_dct = DC.sub(self.required_material, self.activity_material)
```

```python
result = mp.get_plan(requirement_dct = requirement_dct, outcome=True);
```

```python
values = result["values"]
values = {items["name"]: items["value"] for d in values for items in d["items"]}
```

```python
values |= {"基础作战记录" : 200 * 0.0035,
            "初级作战记录" : 400 * 0.0035,
            "中级作战记录" : 1000 * 0.0035,
            "高级作战记录" : 2000 * 0.0035,
            "赤金" : 400 * 0.0035,
            "龙门币" : 0.0035}
```

```python
self.zone_value = dict()
for zone, stages in self.zone_matrix.items():
    self.zone_value[zone] = dict()
    for stage_code, stage_info in stages.items():
        equivalent_sanity = 0
        for item, item_info in stage_info["drop_info"].items():
            if item not in values:
                print(item, item_info)
                continue
            self.zone_value[zone][stage_code] += float(values[item]) * item_info["quantity"] / item_info["times"]
        self.zone_value[zone][stage_code]
```

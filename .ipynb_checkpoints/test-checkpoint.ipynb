```python
headers = {'User-Agent':'ArkPlanner'}
cultivate_url = "https://raw.githubusercontent.com/arkntools/arknights-toolbox/master/src/data/cultivate.json" # evolve, skill
penguin_item_url = "https://penguin-stats.io/PenguinStats/api/v2/items" # itemId, itemName
penguin_stage_url = "https://penguin-stats.io/PenguinStats/api/v2/stages" # stageId, stageCode
operator_id_url = "https://raw.githubusercontent.com/Aceship/AN-EN-Tags/master/json/gamedata/zh_CN/gamedata/excel/character_table.json" # operatorName, operatorId
```

```python
def update_cultivate():
    try:
        req = urllib.request.Request(cultivate_url, None, headers)
        with urllib.request.urlopen(req, timeout=5) as response:
            cultivate = json.loads(response.read().decode())
            with open("./data/cultivate.json", 'w', encoding="utf-8") as fw:
                json.dump(cultivate, fw, indent=2, ensure_ascii=False)
        return True
    except:
        print("Fail to update cultivate.")
        return False

def load_cultivate():
    with open("./data/cultivate.json", 'r', encoding="utf-8-sig") as fr:
        return json.loads(fr.read()) 


def update_item():
    try:
        req = urllib.request.Request(penguin_item_url, None, headers)
        with urllib.request.urlopen(req, timeout=5) as response:
            txt = json.loads(response.read().decode())
            item_id_to_name = ddict()
            for item in txt:
                item_id_to_name[item["itemId"]] = item["name"]
            with open("./data/item_id_to_name.json", 'w', encoding="utf-8") as fw:
                json.dump(item_id_to_name, fw, indent=2, ensure_ascii=False)
        return True
    except:
        print("Fail to update itemId_to_itemName.")
        return False
    
def load_item():
    with open("./data/item_id_to_name.json", 'r', encoding="utf-8-sig") as fr:
        return json.loads(fr.read())
    

def update_operator():
    print("Please update operator / operator_order manually.")
    
    # ==================================
    try:
        req = urllib.request.Request(operator_id_url, None, headers)
        with urllib.request.urlopen(req, timeout=5) as response:
            txt = json.loads(response.read().decode())
        operator_id_to_name = dict()
        for k, v in txt.items():
            operator_id_to_name[k[5:]] = v["name"]
        with open('./data/operator_id_to_name.json', 'w', encoding='utf-8-sig') as f:
            json.dump(operator_id_to_name, f, ensure_ascii=False, indent=2)
    except:
        print("Fail to update operator_id_to_name.")
        return False    
    return False

def load_operator():
    with open("./data/operator_id_to_name.json", 'r', encoding="utf-8-sig") as fr:
        return json.loads(fr.read())
    
    
def load_operator_order():
    with open("./data/operator_order.txt", 'r', encoding="utf-8-sig") as fr:
        return fr.read().split("\n")
    

def update_stage():
    try:
        req = urllib.request.Request(penguin_stage_url, None, headers)
        with urllib.request.urlopen(req, timeout=5) as response:
            txt = json.loads(response.read().decode())
            stage_name_to_id = dict()
            for stage in txt:
                s = ""
                if stage["stageId"][-3:] == 'rep':
                    stage["code"] += '_rep'
                if stage["stageId"][-4:] == 'perm':
                    stage["code"] += '_perm'
                stage_name_to_id[stage["code"]] = stage["stageId"]
            with open("./data/stage_name_to_id.json", 'w', encoding="utf-8") as fw:
                json.dump(stage_name_to_id, fw, indent=2, ensure_ascii=False)
        return True
    except:
        print("Fail to update stageCode_to_stageId.")
        return False
    
def load_stage():
    with open("./data/stage_name_to_id.json", 'r', encoding="utf-8-sig") as fr:
        return json.loads(fr.read())
```

```python
from collections import defaultdict as ddict
```

```python
# calculate required items > required.txt
update = False
if update:
    update_cultivate()
    update_item()
    update_operator()
    update_stage()

cultivate = load_cultivate()
item_id_to_name = load_item()
operator_id_to_name = load_operator()
operator_order = load_operator_order()
stage_name_to_id = load_stage()


operator_order_rv = dict()
for i, v in enumerate(operator_order):
    operator_order_rv[v] = i
depriciate_rate = 1
star_decay_rate = [1, 1, 1, 1, 1,1 ,1, 1]
    
required_items = ddict(lambda : 0)
for operatorId, op in cultivate.items():
    if operatorId in operator_id_to_name and operator_id_to_name[operatorId] in operator_order_rv:
        multiplier = depriciate_rate ** operator_order_rv[operator_id_to_name[operatorId]]
        if len(op["skills"]["elite"]) == 2:
            multiplier *= star_decay_rate[5]
        elif len(op["skills"]["elite"]) == 0:
            multiplier *= star_decay_rate[3]
    else:
        print(operatorId)
        
    for evolve_cost in op["evolve"]:
        for item, cost in evolve_cost.items():
            required_items[item] += cost*multiplier
    for skill_cost in op["skills"]["normal"]: # normal skill
        for item, cost in skill_cost.items():
            required_items[item] += cost*multiplier
    for elite_skill in op["skills"]["elite"]:
        for skill_cost in elite_skill["cost"]:
            for item, cost in skill_cost.items():
                required_items[item] += cost*multiplier
    for uniequip in op["uniequip"]:
        for item_dict in uniequip["cost"]:
            for item, cost in item_dict.items():
                required_items[item] += cost*multiplier
            
required_dct = {item_id_to_name[k] : v for k, v in required_items.items()}
```
